```python
import pandas as pd
from google.colab import files

# Solicitar la carga del archivo CSV
print("Por favor, sube el archivo CSV")
uploaded = files.upload()

# Obtener el nombre del archivo subido
file_path = list(uploaded.keys())[0]

# Leer el archivo ignorando la primera fila y eliminando los caracteres ;;;
df = pd.read_csv(file_path, skiprows=1, sep=',', engine='python', skipfooter=0, on_bad_lines='skip', names=["Brand", "Category", "Color", "Size", "Material", "Price"])

# Mostrar la cantidad de filas antes de la limpieza
initial_rows = df.shape[0]

# 1. Eliminación de valores nulos
df.dropna(inplace=True)
cleaned_rows_na = df.shape[0]

# 2. Eliminación de duplicados
df.drop_duplicates(inplace=True)
cleaned_rows_duplicates = df.shape[0]

# Mostrar resultados
print(f"Filas iniciales: {initial_rows}")
print(f"Filas después de eliminar valores nulos: {cleaned_rows_na}")
print(f"Filas después de eliminar duplicados: {cleaned_rows_duplicates}")

# Visualizar las primeras filas para verificar
print("\nDatos limpios:")
print(df.head())
```
